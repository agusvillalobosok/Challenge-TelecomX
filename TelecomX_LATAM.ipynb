````python
6### GrÃ¡fico de torta DistribuciÃ³n de EVASIÃ“N en GÃ‰NERO, JUBILADO, HIJOS Y PAREJA. ###

def autopct_func(pct, all_vals):
    absolute = int(round(pct / 100. * sum(all_vals)))
    return f'{pct:.1f}%\n({absolute})'

def gen_graf_evasion_GJHP(colormap):
  # Conteos
  genero_counts = df_evasion["GÃ©nero"].value_counts()
  jubilado_counts = df_evasion["Jubilado"].value_counts()
  pareja_counts = df_evasion["Pareja"].value_counts()
  hijos_counts = df_evasion["Hijos"].value_counts()

  # Colores
  num_categorias = len(genero_counts)
  colormap = plt.get_cmap(colormap)  # o 'Pastel1', 'tab20', 'Accent', etc.
  colores = colormap(np.linspace(0, 1, num_categorias))



  # Subplots
  fig, axes = plt.subplots(2, 2, figsize=(12, 7))
  axes = axes.flatten()

  # GrÃ¡fico 1: GÃ©nero
  axes[0].pie(
      genero_counts.values,
      labels=genero_counts.index,
      autopct=lambda pct: autopct_func(pct, genero_counts.values),
      startangle=90,
      colors=colores
  )
  axes[0].set_title('DistribuciÃ³n de GÃ©nero\npara Clientes con Baja', fontfamily="sans-serif", fontsize=16)

  # GrÃ¡fico 2: Jubilado
  axes[1].pie(
      jubilado_counts.values,
      labels=jubilado_counts.index,
      autopct=lambda pct: autopct_func(pct, jubilado_counts.values),
      startangle=90,
      colors=colores
  )
  axes[1].set_title('DistribuciÃ³n de Jubilado\npara Clientes con Baja', fontfamily="sans-serif", fontsize=16)

  # GrÃ¡fico 3: Pareja
  axes[2].pie(
      pareja_counts.values,
      labels=pareja_counts.index,
      autopct=lambda pct: autopct_func(pct, pareja_counts.values),
      startangle=90,
      colors=colores
  )
  axes[2].set_title('DistribuciÃ³n de Pareja\npara Clientes con Baja', fontfamily="sans-serif", fontsize=16)

  # GrÃ¡fico 4: Hijos
  axes[3].pie(
      hijos_counts.values,
      labels=hijos_counts.index,
      autopct=lambda pct: autopct_func(pct, hijos_counts.values),
      startangle=90,
      colors=colores
  )
  axes[3].set_title('DistribuciÃ³n de Hijos\npara Clientes con Baja', fontfamily="sans-serif", fontsize=16)


  # TÃ­tulo inferior
  fig.text(0.5, 0.01, '0 = No es/No tiene\n\n1 = SÃ­ es/SÃ­ tiene', ha='center', fontsize=12)

  # Mostrar grÃ¡fico
  plt.show()
````

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def gen_graf_gastado_barras():

    clie_total_tarifas['Total_Tarifas_Cliente'] = pd.to_numeric(
        clie_total_tarifas['Total_Tarifas_Cliente'], errors='coerce'
    )

    # Eliminar valores no vÃ¡lidos
    clie_total_tarifas.dropna(subset=['Total_Tarifas_Cliente'], inplace=True)


    max_y = clie_total_tarifas["Total_Tarifas_Cliente"].max()
    yticks = np.arange(0, max_y + 1000, 1000)


    plt.figure(figsize=(12, 7))
    plt.bar(clie_total_tarifas.index, clie_total_tarifas["Total_Tarifas_Cliente"],
            color="green", label="Total Tarifas Cliente")


    plt.yticks(yticks)
    plt.grid(True, axis='y')
    plt.xlabel("Usuarios activos")
    plt.ylabel("Total Tarifas Cliente")
    plt.title("Total Tarifas Cliente - GrÃ¡fico de Barras")
    plt.legend()
    plt.tight_layout()
    plt.savefig("/content/graf-gasto-barras.png", transparent=False, dpi=300, bbox_inches="tight")
    plt.show()
````

````python
def gen_graf_gastado3():

    eva_total_tarifas['Total_Tarifas_Cliente'] = pd.to_numeric(
        eva_total_tarifas['Total_Tarifas_Cliente'], errors='coerce'
    )
    clie_total_tarifas['Total_Tarifas_Cliente'] = pd.to_numeric(
        clie_total_tarifas['Total_Tarifas_Cliente'], errors='coerce'
    )


    eva_total_tarifas.dropna(subset=['Total_Tarifas_Cliente'], inplace=True)
    clie_total_tarifas.dropna(subset=['Total_Tarifas_Cliente'], inplace=True)


    max_y = max(
        eva_total_tarifas["Total_Tarifas_Cliente"].max(),
        clie_total_tarifas["Total_Tarifas_Cliente"].max()
    )


    yticks = np.arange(0, max_y + 1000, 1000)


    plt.figure(figsize=(12, 7))
    plt.bar(
        eva_total_tarifas.index,
        eva_total_tarifas["Total_Tarifas_Cliente"],
        label="EvasiÃ³n",
        color="red"
    )


    plt.yticks(yticks)
    plt.grid(True, axis='y')
    plt.xlabel("Cantidad de evasores")
    plt.ylabel("Total Tarifas Cliente")
    plt.title("ComparaciÃ³n de Total Tarifas Cliente")
    plt.legend()
    plt.tight_layout()
    plt.savefig("/content/graf-gasto-barras3.png", transparent=False, dpi=300, bbox_inches="tight")
    plt.show()
````

````python
def gen_graf_curva():

  # Crear listas de valores estadÃ­sticos
  lista_clientes = [
      clie_total_tarifas.describe().loc["min", "Total_Tarifas_Cliente"],
      clie_total_tarifas.describe().loc["25%", "Total_Tarifas_Cliente"],
      clie_total_tarifas.describe().loc["50%", "Total_Tarifas_Cliente"],
      clie_total_tarifas.describe().loc["75%", "Total_Tarifas_Cliente"],
      clie_total_tarifas.describe().loc["max", "Total_Tarifas_Cliente"],
  ]

  lista_eva = [
      eva_total_tarifas.describe().loc["min", "Total_Tarifas_Cliente"],
      eva_total_tarifas.describe().loc["25%", "Total_Tarifas_Cliente"],
      eva_total_tarifas.describe().loc["50%", "Total_Tarifas_Cliente"],
      eva_total_tarifas.describe().loc["75%", "Total_Tarifas_Cliente"],
      eva_total_tarifas.describe().loc["max", "Total_Tarifas_Cliente"],
  ]


  plt.figure(figsize=(10, 6))
  plt.plot(lista_eva, marker="o", label="EvasiÃ³n")
  plt.plot(lista_clientes, marker="o", label="Clientes")

  plt.title("Curva: tarifa total de clientes y evasores")
  plt.xticks(range(5), ["min", "25%", "50%", "75%", "max"])
  plt.legend()

  # Anotar valores sobre los puntos
  for i, val in enumerate(lista_eva):
      plt.text(i, val + max(lista_eva)*0.02, f"{val:,.0f}", ha='center', color='red')
  for i, val in enumerate(lista_clientes):
      plt.text(i, val + max(lista_clientes)*0.02, f"{val:,.0f}", ha='center', color='green')

  plt.grid(True, axis='y')
  plt.tight_layout()
  plt.savefig("/content/graf-curva.png", transparent=False, dpi=300, bbox_inches="tight")
  plt.show()

````
